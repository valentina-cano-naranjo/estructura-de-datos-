## 1
Crear un sistema simple de gestión de biblioteca que permita agregar libros, prestar libros a usuarios, y devolver libros. El sistema deberá manejar dos clases principales: Libro y Biblioteca.
Requisitos:
Clase Libro:

    Atributos:
        titulo (str): El título del libro.
        autor (str): El autor del libro.
        isbn (str): El número ISBN del libro.
        disponible (bool): Indica si el libro está disponible para préstamo.

    Métodos:
        __init__(self, titulo, autor, isbn): Constructor para inicializar un nuevo libro. El estado inicial de disponible debe ser True.
        prestar(self): Cambia el estado de disponible a False.
        devolver(self): Cambia el estado de disponible a True.
        informacion(self): Devuelve una cadena con la información del libro, incluyendo si está disponible o no.

Clase Biblioteca:

    Atributos:
        nombre (str): El nombre de la biblioteca.
        libros (list): Una lista que almacena los libros de la biblioteca.

    Métodos:
        __init__(self, nombre): Constructor para inicializar una nueva biblioteca con un nombre y una lista vacía de libros.
        agregar_libro(self, libro): Añade un nuevo libro a la biblioteca.
        mostrar_libros(self): Muestra todos los libros disponibles en la biblioteca.
        prestar_libro(self, isbn): Busca un libro por su ISBN y si está disponible, lo presta.
        devolver_libro(self, isbn): Busca un libro por su ISBN y lo devuelve si está prestado.

## 2
Crear un sistema de gestión de matrículas universitarias que maneje diferentes tipos de estudiantes (por ejemplo, estudiantes de pregrado y posgrado). El sistema debe permitir calcular la matrícula dependiendo del tipo de estudiante y las materias inscritas. Se utilizarán clases base y derivadas para representar los diferentes tipos de estudiantes.
Requisitos:
Clase Base Estudiante:

    Atributos:
        nombre (str): El nombre del estudiante.
        id_estudiante (str): El ID único del estudiante.
        materias (list): Una lista de materias inscritas por el estudiante.

    Métodos:
        __init__(self, nombre, id_estudiante): Constructor para inicializar un estudiante con su nombre e ID.
        inscribir_materia(self, materia): Añade una materia a la lista de materias inscritas.
        calcular_matricula(self): Método que será sobrescrito en las clases derivadas para calcular la matrícula dependiendo del tipo de estudiante.

Clase Derivada EstudiantePregrado (hereda de Estudiante):

    Atributos:
        descuento_beca (float): Porcentaje de descuento en la matrícula debido a una beca.

    Métodos:
        __init__(self, nombre, id_estudiante, descuento_beca): Constructor para inicializar un estudiante de pregrado con un descuento en la matrícula.
        calcular_matricula(self): Calcula la matrícula considerando el número de materias inscritas y el descuento por beca.

Clase Derivada EstudiantePosgrado (hereda de Estudiante):

    Atributos:
        costo_por_materia (float): Costo fijo por cada materia inscrita.

    Métodos:
        __init__(self, nombre, id_estudiante, costo_por_materia): Constructor para inicializar un estudiante de posgrado con un costo por materia.
        calcular_matricula(self): Calcula la matrícula basada en el costo por materia y el número de materias inscritas.

## 3
Crear un sistema de gestión de despegues en un aeropuerto que maneje diferentes tipos de aviones (por ejemplo, aviones comerciales y aviones privados). El sistema debe permitir gestionar la cola de despegues, verificar las condiciones de cada avión antes del despegue, y organizar el orden de despegue según la prioridad de los aviones.
Requisitos:
Clase Base Avion:

    Atributos:
        modelo (str): El modelo del avión.
        matricula (str): La matrícula del avión.
        combustible (int): Nivel de combustible en porcentaje.
        prioridad (int): La prioridad del avión para el despegue (1 es la más alta).

    Métodos:
        __init__(self, modelo, matricula, combustible, prioridad): Constructor para inicializar un avión con su modelo, matrícula, nivel de combustible y prioridad.
        verificar_condiciones(self): Verifica si el avión está listo para el despegue (debe tener más del 30% de combustible).
        despegar(self): Realiza el proceso de despegue (simulación).

Clase Derivada AvionComercial (hereda de Avion):

    Atributos:
        pasajeros (int): Número de pasajeros a bordo.

    Métodos:
        __init__(self, modelo, matricula, combustible, prioridad, pasajeros): Constructor para inicializar un avión comercial.
        verificar_condiciones(self): Verifica si el avión comercial está listo para el despegue (el número de pasajeros debe ser menor que la capacidad máxima y debe tener suficiente combustible).
        despegar(self): Realiza el proceso de despegue mostrando detalles específicos del avión comercial.

Clase Derivada AvionPrivado (hereda de Avion):

    Atributos:
        dueno (str): El nombre del dueño del avión privado.

    Métodos:
        __init__(self, modelo, matricula, combustible, prioridad, dueno): Constructor para inicializar un avión privado.
        verificar_condiciones(self): Verifica si el avión privado está listo para el despegue (debe tener suficiente combustible y estar en buenas condiciones).
        despegar(self): Realiza el proceso de despegue mostrando detalles específicos del avión privado.

Clase Aeropuerto:

    Atributos:
        nombre (str): El nombre del aeropuerto.
        cola_despegue (list): Una lista que almacena los aviones en cola para el despegue.

    Métodos:
        __init__(self, nombre): Constructor para inicializar un aeropuerto con su nombre y una cola vacía.
        agregar_a_cola(self, avion): Añade un avión a la cola de despegue.
        organizar_cola(self): Organiza la cola de despegue en función de la prioridad de los aviones.
        procesar_despegues(self): Procesa los despegues de los aviones en la cola, verificando sus condiciones y despegando en orden de prioridad.

## 4
Desarrolla un sistema de gestión para un taller de motos. Este sistema debe manejar diferentes tipos de motos, las reparaciones necesarias, la gestión de mecánicos, y la asignación de trabajos. El sistema debe permitir registrar diferentes tipos de motos (por ejemplo, motos de calle y motos de carreras), gestionar las reparaciones, y coordinar el trabajo de los mecánicos.
Requisitos:
Clase Base Moto:

    Atributos:
        marca (str): La marca de la moto.
        modelo (str): El modelo de la moto.
        ano (int): Año de fabricación.
        kilometraje (int): Kilometraje de la moto.
        estado (str): El estado actual de la moto (por ejemplo, "En reparación", "Lista para entregar").

    Métodos:
        __init__(self, marca, modelo, ano, kilometraje): Constructor para inicializar una moto.
        detalles(self): Muestra los detalles de la moto.
        cambiar_estado(self, nuevo_estado): Cambia el estado de la moto.

Clase Derivada MotoCalle (hereda de Moto):

    Atributos:
        cilindraje (int): Cilindraje de la moto.

    Métodos:
        __init__(self, marca, modelo, ano, kilometraje, cilindraje): Constructor para inicializar una moto de calle.
        detalles(self): Muestra detalles específicos de la moto de calle.

Clase Derivada MotoCarreras (hereda de Moto):

    Atributos:
        velocidad_maxima (int): Velocidad máxima alcanzada por la moto.
        patrocinador (str): El patrocinador principal de la moto.

    Métodos:
        __init__(self, marca, modelo, ano, kilometraje, velocidad_maxima, patrocinador): Constructor para inicializar una moto de carreras.
        detalles(self): Muestra detalles específicos de la moto de carreras.

Clase Mecanico:

    Atributos:
        nombre (str): Nombre del mecánico.
        especialidad (str): Especialidad del mecánico (por ejemplo, "Motor", "Suspensión").
        motos_asignadas (list): Lista de motos que le han sido asignadas al mecánico.

    Métodos:
        __init__(self, nombre, especialidad): Constructor para inicializar un mecánico.
        asignar_moto(self, moto): Asigna una moto al mecánico para su reparación.
        reparar_moto(self, moto): Simula la reparación de una moto y cambia su estado a "Reparada".

Clase Taller:

    Atributos:
        nombre (str): Nombre del taller.
        motos_en_reparacion (list): Lista de motos que están en el taller para reparación.
        mecanicos (list): Lista de mecánicos que trabajan en el taller.

    Métodos:
        __init__(self, nombre): Constructor para inicializar un taller.
        registrar_moto(self, moto): Registra una moto en el taller.
        asignar_mecanico(self, moto, mecanico): Asigna un mecánico a una moto específica.
        reparar_motos(self): Procesa la reparación de todas las motos en el taller.

## 5

En este ejercicio, diseñarás un sistema avanzado de gestión para un restaurante. El sistema manejará la gestión de mesas, reservas, menús, pedidos, inventarios de ingredientes, personal (cocineros y meseros), facturación, y reportes de ventas. Incluirás conceptos avanzados de POO como herencia, polimorfismo, composición, agregación, manejo de excepciones, uso de propiedades (@property), métodos estáticos y de clase, y patrones de diseño como Singleton y Factory.

Requisitos:

Clase Cliente:

    Atributos:
        nombre (str): Nombre del cliente.
        telefono (str): Teléfono de contacto del cliente.
        reservas (list): Lista de reservas realizadas por el cliente.

    Métodos:
        __init__(self, nombre, telefono): Constructor para inicializar un cliente.
        hacer_reserva(self, reserva): Añade una reserva a la lista de reservas del cliente.
        detalles(self): Muestra los detalles del cliente y sus reservas.

Clase Mesa:

    Atributos:
        numero (int): Número de la mesa.
        capacidad (int): Capacidad máxima de la mesa.
        reservada (bool): Indicador de si la mesa está reservada.

    Métodos:
        __init__(self, numero, capacidad): Constructor para inicializar una mesa.
        reservar(self): Marca la mesa como reservada.
        liberar(self): Libera la mesa.
        detalles(self): Muestra los detalles de la mesa.

Clase Reserva (Composición con Mesa y Cliente):

    Atributos:
        cliente (Cliente): Cliente que realiza la reserva.
        mesa (Mesa): Mesa reservada.
        fecha_hora (datetime): Fecha y hora de la reserva.

    Métodos:
        __init__(self, cliente, mesa, fecha_hora): Constructor para inicializar una reserva.
        detalles(self): Muestra los detalles de la reserva.

Clase Menu:

    Atributos:
        items (list): Lista de platos disponibles en el menú.

    Métodos:
        __init__(self): Constructor para inicializar un menú vacío.
        agregar_item(self, item): Añade un plato al menú.
        remover_item(self, item): Remueve un plato del menú.
        detalles(self): Muestra todos los platos del menú.

Clase Plato:

    Atributos:
        nombre (str): Nombre del plato.
        precio (float): Precio del plato.
        ingredientes (list): Lista de ingredientes necesarios para preparar el plato.

    Métodos:
        __init__(self, nombre, precio, ingredientes): Constructor para inicializar un plato.
        detalles(self): Muestra los detalles del plato.

Clase InventarioIngredientes (Singleton):

    Atributos:
        ingredientes (dict): Diccionario de ingredientes disponibles, con el nombre del ingrediente como clave y la cantidad como valor.

    Métodos:
        __new__(cls): Asegura que solo haya una instancia de esta clase.
        agregar_ingrediente(self, nombre_ingrediente, cantidad): Añade ingredientes al inventario.
        usar_ingrediente(self, nombre_ingrediente, cantidad): Usa ingredientes del inventario, verificando la disponibilidad.

Clase Pedido (Composición con Plato y Mesa):

    Atributos:
        mesa (Mesa): Mesa que realizó el pedido.
        platos (list): Lista de platos pedidos.
        estado (str): Estado del pedido ("En preparación", "Listo", "Entregado").

    Métodos:
        __init__(self, mesa): Constructor para inicializar un pedido.
        agregar_plato(self, plato): Añade un plato al pedido.
        cambiar_estado(self, nuevo_estado): Cambia el estado del pedido.
        detalles(self): Muestra los detalles del pedido.

Clase Personal (Base para Cocinero y Mesero):

    Atributos:
        nombre (str): Nombre del empleado.
        puesto (str): Puesto del empleado.
        turno (str): Turno de trabajo del empleado.

    Métodos:
        __init__(self, nombre, puesto, turno): Constructor para inicializar un empleado.
        detalles(self): Muestra los detalles del empleado.

Clase Cocinero (Hereda de Personal):

    Atributos:
        especialidad (str): Especialidad del cocinero.

    Métodos:
        __init__(self, nombre, turno, especialidad): Constructor para inicializar un cocinero.
        preparar_pedido(self, pedido): Cambia el estado del pedido a "Listo" cuando se completa la preparación.

Clase Mesero (Hereda de Personal):

    Atributos:
        mesas_asignadas (list): Lista de mesas asignadas al mesero.

    Métodos:
        __init__(self, nombre, turno): Constructor para inicializar un mesero.
        asignar_mesa(self, mesa): Asigna una mesa al mesero.
        entregar_pedido(self, pedido): Cambia el estado del pedido a "Entregado".

Clase Factura:

    Atributos:
        pedido (Pedido): Pedido asociado a la factura.
        total (float): Total de la factura.

    Métodos:
        __init__(self, pedido): Constructor para inicializar una factura.
        calcular_total(self): Calcula el total de la factura sumando los precios de los platos.
        generar_factura(self): Genera la factura mostrando el desglose de platos y el total.

Clase Restaurante (Agregación con Cliente, Mesa, Personal, InventarioIngredientes):

    Atributos:
        nombre (str): Nombre del restaurante.
        clientes (list): Lista de clientes.
        mesas (list): Lista de mesas.
        personal (list): Lista de empleados.
        menu (Menu): Menú del restaurante.
        inventario (InventarioIngredientes): Instancia del inventario de ingredientes.

    Métodos:
        __init__(self, nombre): Constructor para inicializar un restaurante.
        registrar_cliente(self, cliente): Registra un cliente en el restaurante.
        agregar_mesa(self, mesa): Añade una mesa al restaurante.
        agregar_empleado(self, empleado): Añade un empleado al restaurante.
        procesar_pedido(self, mesa, platos): Procesa un pedido asignándolo a un cocinero disponible y luego a un mesero para su entrega.
        generar_factura(self, pedido): Genera y muestra la factura para un pedido.
