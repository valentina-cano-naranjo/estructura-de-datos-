## Clase
Modelo donde se contienen las caracteristicas comunes de un grupo de objetos, está compuesta por atributos y metodos.
Para crear una clase se utiliza la palabra reservada class y a continuación el nombre de la clase. La definición de la clase se pone entre las llaves de apertura y cierre. El nombre de la clase empieza por letra mayúscula.

```python
class Carro:
    atrinbutodeclass = 0 #atributo de clase
    def __init__(self, marca, modelo, anio, cilindraje):
        
        self.marca = marca #ateriobutos de instancia
        self.modelo = modelo
        self.ano = anio
        self.odometro = 0

    def descricao(self):
        
        
        self.atrinbutodeclass = 1
        print(f"Esse carro é um {self.anio} {self.marca} {self.modelo}.") #acceso al atributo de instancia

    def le_odometro(self):
        
        print(f"Esse carro tem {self.odometro} km rodados.")

    def atualiza_odometro(self, km):
        if km >= self.odometro:
            self.odometro = km
        else:
            print("Você não pode diminuir o odômetro.")

    def incrementa_odometro(self, km):
        self.odometro += km
```

# Atributos
Los atributos son las caracterísiticas individuales que diferencian un objeto de otro y determinan su apariencia, estado u otras cualidades. Los atributos se guardan en variables denominadas de instancia, y cada objeto particular puede tener valores distintos para estas variables.

Las variables de instancia también denominados miembros dato, son declaradas en la clase pero sus valores son fijados y cambiados en el objeto.

# Los constructores
Un objeto de una clase se crea llamando a una función especial denominada constructor de la clase. El constructor se llama de forma automática cuando se crea un objeto, para situarlo en memoria e inicializar los miembros dato declarados en la clase. El constructor tiene el mismo nombre que la clase. Lo específico del constructor es que no tiene tipo de retorno.

# Los objetos


```python
class Carro:
  def __init__(self,color,largo):
    self.color = color
    self.tanqueCom = 10 
    self.largo = largo
    self.numLlantas = 4
    self.velocidad = 0
    self.encendido = False

  def start(self):
    self.encendido = True
    return "el carro está encendido"
    
  def apagado(self):
    self.encendido = False
    self.velocidad = 0
    print("el carro está apagado")
    
  
  def acelerar(self,v):
    if v < 0:
      print("no puede velocidades negaticvcas)")
    else:
      
        if self.encendido:
          self.velocidad += v
          print("El carro va a: ", self.velocidad, "Km/h")
        else:
          print("Prenda el carro")
  
  def frenar(self):
    if self.velocidad >= 0:
      self.velocidad = 0
      print("El carro está detenido")
    
  def desacelerar(self,v):
      if self.encendido:
        self.velocidad -= v
        if self.velocidad <= 0:
          self.velocidad = 0
          print("el carro esta detemnido")
        else:
          print("El carro va a: ", self.velocidad, "Km/h")
            
        
      else:
        print("Prenda el carro")







```
